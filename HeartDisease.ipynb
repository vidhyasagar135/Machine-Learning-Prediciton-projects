```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
ds=pd.read_csv('/content/heart_disease_data.csv')
```

```python
ds.head()
```

```python
X=ds.drop(columns='target',axis=1)
Y=ds['target']
```

```python
X_train , X_test ,Y_train ,Y_test =train_test_split(X,Y,test_size=0.2,stratify=Y)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
model =LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
X_train_pred=model.predict(X_train)
acc_score=accuracy_score(X_train_pred,Y_train)
print(acc_score)
```

```python
X_test_pred=model.predict(X_test)
test_acc_score=accuracy_score(X_test_pred,Y_test)
print(test_acc_score)
```

```python
input_data =(52,0,2,136,196,0,0,169,0,0.1,1,0,2)
np_data =np.asarray(input_data)
reshape_data =np_data.reshape(1,-1)
prediction=model.predict(reshape_data)
print(prediction)
if(prediction[0]==0):
  print("No Heart Disease")
else:
  print("Heart Disease")
```
